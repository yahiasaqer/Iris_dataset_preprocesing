```python
import pandas as pd
data = pd.read_csv('iris_mod.csv')
data.head()
```

```python
import matplotlib.pyplot as plt
d = data
d.hist()
plt.show()
```

```python
d.drop('mo', axis = 1, inplace = True) #you can delete more than one
d.head(3)
```

```python
list = ['a1', 'a2', 'a3', 'a4', 'a5', 'Label']
d.drop_duplicates(subset=list, keep = 'first', inplace = True) # keep can be "last" ,True or False.
d.head(10)
```

```python
d.info()
```

```python
d.describe()
```

```python
missing = pd.DataFrame({'missing': d.isnull().sum()}) #finding null values
missing
```

```python
#Filling null values

d['a1'] = d['a1'].fillna(d['a1'].median())
d['a3'] = d['a3'].fillna(d['a3'].mean())

missing = pd.DataFrame({'missing': d.isnull().sum()}) #finding null values
missing
```

```python
d.describe()
```

```python
atts = list = ['a1', 'a2', 'a3', 'a4', 'a5']
plt.figure(figsize= (20, 10))
d[atts].boxplot()
plt.title('iris attributes')
plt.show()
```
